```python
voice=pd.DataFrame(df[["MDVP:Fo(Hz)","MDVP:Fhi(Hz)","MDVP:Flo(Hz)","status"]])
frequency=pd.DataFrame(df[["MDVP:Jitter(%)","MDVP:Jitter(Abs)","MDVP:RAP","MDVP:PPQ","Jitter:DDP","status"]])
amptitude=pd.DataFrame(df[["MDVP:Shimmer","MDVP:Shimmer(dB)","Shimmer:APQ3","Shimmer:APQ5","MDVP:APQ","Shimmer:DDA","status"]])
tone=pd.DataFrame(df[["NHR","HNR","status"]])
variation=pd.DataFrame(df[["spread1","spread2","PPE","RPDE","D2","DFA","status"]])
voice1=pd.DataFrame(df[["status"]])

x=voice.values
clf = IsolationForest(contamination=.09)
clf.fit(x)
predictions = clf.predict(x)
(predictions<0).mean()
abn_ind = np.where(predictions < 0)
voice=voice.loc[abn_ind]
```

Linear Regression

```python
x=np.array(voice["MDVP:Fhi(Hz)"]).reshape(-1,1)
y=np.array(voice["MDVP:Flo(Hz)"])
model = LinearRegression()
model.fit(x, y)
model.score(x, y)
```

```python
y_pred = model.intercept_ + model.coef_ * x
plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='red')
plt.show()
```

Multiple Linear Regression

```python
x=frequency.values
clf = IsolationForest(contamination=.09)
clf.fit(x)
predictions = clf.predict(x)
(predictions<0).mean()
abn_ind = np.where(predictions < 0)
frequency=frequency.loc[abn_ind]
x=amptitude.values
clf = IsolationForest(contamination=.09)
clf.fit(x)
predictions = clf.predict(x)
(predictions<0).mean()
abn_ind = np.where(predictions < 0)
amptitude=amptitude.loc[abn_ind]
x=np.array(frequency["MDVP:Jitter(%)"],frequency["MDVP:Jitter(Abs)"]).reshape(-1,1)
y=np.array(amptitude["MDVP:Shimmer"])
model = LinearRegression()
model.fit(x, y)
model.score(x, y)
```

```python
y_pred = model.intercept_ + np.sum(model.coef_ * x, axis=1)
plt.scatter(x,y,color='blue')
plt.plot(x,y_pred,color='red')
plt.show()
```
